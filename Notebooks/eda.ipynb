```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('../dados/victor.csv', parse_dates=['data'], dayfirst=True)
```

```python
df.head()
```

```python
fig = px.line(df.groupby(['data', 'loja']).sum().reset_index(), x='data', y='qtde', title='Vendas ao longo do tempo', color='loja')
```

```python
fig.show()
```

```python
fig = px.line(df.groupby(['data','tipoCliente']).sum().reset_index(), x='data', y='qtde', title='Vendas ao longo do tempo', color='tipoCliente')
```

```python
fig.show()
```
